## OpenML-CC18 Joy Plot
Downloads all available model evaluations on the CC18 benchmark suite tasks and visualizes them as a Joy Plot

```python
%matplotlib inline
import openml
import seaborn as sns
import numpy as np
import pandas as pd
import sys
from scipy.stats import norm
from matplotlib import pyplot
from tqdm import tqdm
```

#### Fetch results
Download all AUC evaluations for all tasks from 'OpenML-CC18' benchmarking suite

* You can repeat this for any other suite of OpenML tasks by replacing the name of the suite
* You can use any evaluation metric you like, here we choose AUC.
* Given the huge amount of results, we're fetching experiments 10,000 at the time

```python
bench = openml.study.get_suite('OpenML-CC18') # obtain the tasks
tasks, aucs = [], []
for i in tqdm(range(400)): # Fetch results 10000 results at a time
  evals = openml.evaluations.list_evaluations(tasks=bench.tasks, function='area_under_roc_curve', 
                                              output_format="dataframe", offset=i*10000, size=10000)
  if evals.shape[0] > 0:
    tasks.extend(evals.data_name.to_list()), aucs.extend(evals.value.to_list())
```

```python
# Number of returned results
len(aucs)
```

#### Organize results
We reformat the results in a dataframe for easy plotting. For the joy plot we only need the task ID and score of each experiment

```python
# Create dataframe and sort by median over individual tasks
df = pd.DataFrame({'score':aucs, 'task':tasks})
df['task'] = pd.Categorical(df['task'], df.groupby(['task']).median().sort_values('score').index.tolist())
df.sort_values('task');
```

#### Side by side plotting

For use in papers, we plot the results in two columns, with the most difficult tasks 
(lowest average score) on the left and the others on the right. This requires adding
a row and column value to the dataframe

```python
tasklist = df["task"].cat.categories.to_list()
```

```python
def leftright(task):
  if tasklist.index(task) < 36:
    return 0;
  else:
    return 1;

def taskrow(task):
  return tasklist.index(task) % 36;
```

```python
df['column'] = df.apply (lambda row: leftright(row.task), axis=1)
df['row'] = df.apply (lambda row: taskrow(row.task), axis=1)
```

#### Create the joy plot.  
Adapted from https://seaborn.pydata.org/examples/kde_joyplot.html

```python
def joyplot(smoothing, aspect, height, overlap):
  sns.set(style="white", rc={"axes.facecolor": (0, 0, 0, 0)}, font_scale=3)
  # Initialize the FacetGrid object
  pal = sns.cubehelix_palette(10, rot=-.25, light=.7)
  g = sns.FacetGrid(df, row="row", col="column", hue="task", aspect=aspect, height=height, ylim=(0, 15))

  # Draw the densities in a few steps
  g.map(sns.kdeplot, "score", clip=(0,30), shade=True, alpha=1, lw=1.5, bw_method=smoothing)
  g.map(sns.kdeplot, "score", clip=(0,30), color="w", lw=2, bw_method=smoothing)
  g.map(pyplot.axhline, y=0, lw=2, clip_on=False)

  # Define and use a simple function to label the plot in axes coordinates
  def label(x, color, label):
      ax = pyplot.gca()
      ax.text(0, .2, label, fontsize=28, fontweight="bold", color=color, 
              ha="left", va="center", transform=ax.transAxes)

  g.map(label, "score")

  # Set the subplots to overlap
  g.fig.subplots_adjust(hspace=-overlap)

  # Remove axes details that don't play will with overlap
  g.set(yticks=[])
  g.set_titles("")
  g.fig.suptitle('AUC score distribution')
  g.fig.subplots_adjust(top=.96)
  g.despine(bottom=True, left=True);
  pyplot.savefig('CC18-joyplot.png', dpi=300)
```

```python
joyplot(0.1, 30, 0.8, 0.45) #smoothing, aspect, height, overlap
```
